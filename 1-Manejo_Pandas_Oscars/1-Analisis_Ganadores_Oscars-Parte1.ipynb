```python
#Ahora si queremos agregar/poner una columna en un lugar especifico.... vamos a querer nuestra nueva columna entre 'date_of_birth' y 'race_ethnicity'....
#Primero vamos a obtener la posición de la columna en la cual queremos agregar (date_of_birth) la nueva columna ('last_ten_years' --> la que antes borramos) -->en el caso de arriba es facil, entre la columna 1 y 2 quisieramos agregar el dato... pero si fueran más columnas se nos complejizaría esto)
#Para obtener la posicion usamos get_loc:
date_of_birth_position=oscars_short.columns.get_loc('date_of_birth')+1 #A esa posicion (1) le sumamos 1 ya que nosotros buscamos agregar una columna DETRAS de esta... entonces le sumamos el 1.
date_of_birth_position #Vemos que está en la posición 2 
```

```python
#Ahora vamos a ver como seleccionar datos USANDO INDICES BOOLEANOS:
#Primero construimos un indice:
oscars_last_20_years=oscars_short['year_of_award']>1999 
#Habiamos visto que si seleccionabamos una columna (arriba seleccionamos year_of_award)... esto nos daria
#una SERIE... y si aplicamos una operación sobre esa columna nos va a dar una serie con el resultado de esa operación.
#Osea que si aplicamos una operación booleana sobre una columna nos va a dar una SERIE booleana que va a tener T o F en función de si cumple o no la condicion que nosotros le estamos poniendo:
#En este caso vamos a armar una serie que se va a llamar oscars_last_20_years y va a tener T si el premio fue otorgado dentro de los ultimos 20 años o  F si fue entregado antes de los 20 años. 
```
