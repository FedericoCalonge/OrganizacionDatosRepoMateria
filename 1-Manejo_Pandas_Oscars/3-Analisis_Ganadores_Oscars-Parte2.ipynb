```python
#Primero recordamos el DF que estamos utilizando:
oscars_short = pd.read_csv('data/Oscars-demographics-DFE.csv',
                        encoding='latin-1',
                        usecols = ['birthplace', 'date_of_birth','race_ethnicity','religion','sexual_orientation','year_of_award','award','movie','person'])
oscars_short.head(10)
```

```python
#Hacemos lo que hicimos anteriormente en 2-Visualizaicon_Analisis_Ganadores_Oscars

oscars_short.loc[:,'date_of_birth'] = pd.to_datetime(oscars_short['date_of_birth'],errors = 'coerce') #date_of_birth es de tipo object en oscars_shorts.. la convertimos a formato fecha (Date):
oscars_short.drop_duplicates(inplace=True)   #Borramos los duplicados.

#Corregimos lo de los nacimientos futuros y fuera de tiempo (teniamos que restarles 100 años):
oscars_short['year_of_birth']=oscars_short.loc[:,'date_of_birth'].dt.year #Primero asignamos el año de nacimiento a 'year_of_birth' (tal cual lo obtuvimos)
oscars_short['age_at_award']=oscars_short['year_of_award']-(oscars_short['date_of_birth'].dt.year)#Ahora en age_at_award guardamos el año premio - año fecha nacimiento (nos va a seguir dando mal igualmente).
oscars_short.loc[oscars_short['age_at_award']<1,'year_of_birth'] = oscars_short.loc[oscars_short['age_at_award']<1,'year_of_birth'] - 100 #--> Entonces.. a year_of_birth... para los que tienen una edad al momento de recibir el premio <1, le RESTAMOS 100. Ahora year_of_birth nos aparecera como debia ser.
oscars_short['age_at_award']=oscars_short['year_of_award']-(oscars_short['year_of_birth']) #Ahora corregimos el age_at_award con los calculos de años correctos.

oscars_short
```

### 1-Split-Apply-Combine y Group By Transform

```python
#Vamos a ver como realizar agrupamientos para:
    #-Aggregation,   
    #Filtration,
    #Transformation.

#El uso más común para groupby es el 1ro (para Aggregation)... donde tomamos muchos valores y los convertimos a un único valor.
#Para hacer esto definiremos:
    #un "__aggregating column__" (de donde tomaremos el valor para hacer la agregación) 
    #y una "__aggregating function__" (cómo hacemos la agregación). 

#Algunas de las funciones de agregación son: _mean, sum, max, min, count_, etc...

#La manera más facil de hacer esto es con GROUP BY + APPLY...
```

```python
#Vamos a ver la cantidad de premios que gano cada persona...
#Para esto agrupamos todos los datos de cada una de las personas, los contabilizamos y luego
#generamos un resultado final indicando para cada persona cuales son los premios que gano. 
#Este split que vamos a realizar por claves lo vamos a manejar con el group by.
#Con el group by podemos agrupar por algún valor (puede ser CUALQUIERA basicamente... podemos ver la ciudad de nacimiento por ej. y ver la cantidad de premios por ciudades o por persona; o sino agrupar por tipo de premio o por año de nacimiento, etc... 
#En conclusión podemos agrupar por el criterio que más nos guste). Y de esta manera sacar ciertas estadisticas sobre ciertos grupos y después vamos a unir todo esto.

#Hacemos nuestro primer groupby... hacemos un group by por "award" (el nombre del premio)
#y vamos a calcular para la columna "age_at_award" el promedio y la cantidad (mean y count).
#Osea lo que vamos a calcular es CUAL ES LA EDAD PROMEDIO (mean) DE LA GENTE QUE GANO CADA UNO DE LOS PREMIOS. Y además hacemos un count para ver CUANTOS ELEMENTOS TENEMOS DE CADA UNO.
grouped=oscars_short.groupby(['award']).agg({'age_at_award':['mean','count']}) 
grouped
```

```python
#De esta manera, arriba vemos que la estructura que nos generó el group by tiene 2 NIVELES...
#Para age_at_Award tenemos mean y count. Y después tenemos award (que es el dato agrupado).
#Vemos que en total (86+86+86+1+78+78) hay 415 elementos.. BIEN, son 416 elementos (Falta uno... ver porque) pero esta bien.
#VVemos que tenemos 86 premios a mejor actor. Y en promedio los ganadores de Best Actor tenian 45 años.

#Tenemos que considerar un mínimos de elementos para considerar cosas... por ej. en un publo chico con 1 asesino vamos a tener una alta tasa de asesinatos en ese pueblo comparado con una ciudad grande que tenga 1 solo asesino (tendrá una tasa muy baja)
#Y entonces tendriamos que decir "si este minimo de población no se cumple entonces no la tenemos en cuenta porque van a influenciar mucho en el resultado las 15 personas que viven en el pueblo"
#Acá es lo mismo, en este caso cada categoría tiene gran cantidad de datos... pero si tuvieramos premios con muy pocos datos eventualmente quisieramos eliminar esto de nuestros analisis porque nos influenciarian con valores que estan tomados prácticamente solos.
#Por esto siempre hacemos el count, para ver eso. Por ej. podemos considerar que hayan más de 10 actores en cada premio para sacar su promedio. Y sino que no lo muestre. --> HACER ESO. 
```

```python
#Ahora hacemos casi lo mismo pero agrupando por año de premio enves de categoria de premio:
grouped=oscars_short.groupby(['year_of_award']).agg({'age_at_award':['mean','count']}) 
grouped
```

```python
#Como vemos arriba hay muchisima más variación que antes.
```

```python
#Ahora sobre el grouped calculado previamente hacemos un FILTER y un SORT:

#Primero para filtrar los que tengan más de 5 count hacemos asi grouped[('age_at_award','count')]>5,:] --> ya que lo podemos ver como una TUPLA... 
#Esta tupla es --> ('age_at_award','count')... donde primero tenemos que ir a "age_at_award" y después a "count" (por la división de 2 niveles dicha anteriormente). De esta manera estamos viendo el count de la columna age_At_award.
#Y luego hacemos obteniendo esas columnas con count>5 hacemos un SORT para ordenar de mayor a menor sobre ('age_at_award','mean') --> TUPLA.... osea el campo mean de age_at_award. 
#Y nos quedamos con las primeras 10 (head(10)): hacemos un "top ten"
grouped.loc[grouped[('age_at_award','count')]>5,:].sort_values(('age_at_award','mean'),ascending=False).head(10)

#Entonces vimos que obtuvimos un listado con los años en donde se entregaron mayores premios y sus porcentajes de edades.

#De esta manera aplicamos una agrupación, luego filtramos y luego ordenamos. Esto es MUY COMÚN y lo vamos a utilizar muchas veces.
```

```python
#Ahora vamos a hacer un bar plot... vamos a tomar un group by por award (las categorias: mejor actor, actriz, etc.)
#y a eso le calculamos el promedio (.mean()).. osea dividimos nuestro df en los 5 premios y para cada uno aplicamos el mean()
groupedPromedioAward=oscars_short.groupby('award').mean() #Promedio solo de las variables numericas (en este caso solo tenemos 3)
groupedPromedioAward
```

```python
#Ahora vamos a querer armar un DF (porque antes NO teniamos un DF, teniamos una estructura donde teniamos un multiindice y la clave eran los valores por los cuales agrupabamos)
#Por ej. si hacemos un group by por award y religion tanto award como religion nos quedarán como claves de la estructura que armamos.
#Entonces si hacemos un size nos quedará la estructura con esas clave y 1 único valor (el count de religion).
#Si queremos entonces volver al DF con su estrcutrua "plana" tenemos que hacer un reset_index(). Esto nos creará un indice 0 que tendrá
#los count de religion. Y por esto renombramos esta columna (0) para que nos quede "count_religion".

#Entonces, 1ro hacemos el size:
oscars_short_by_religion=oscars_short.groupby(['award','religion']).size()
oscars_short_by_religion
```

```python
#Anteriormente vimos que podiamos aplicar más de una operación al mismo tiempo con el aggregate (.AGG)...
#Si queremos solo el promedio le dejamos el mean, pero si queremos aplicar más operaciones tenemos que usar la estrcutra de abajo:
#(donde age_At_award es la columna en la cual aplicamos el mean y el count).

groupedYearAndAge_Award=oscars_short.groupby(['year_of_award']).agg({'age_at_award':['mean','count']}) 
groupedYearAndAge_Award.head(5)
```

```python
#Ahora aplicamos diferentes operaciones a diferentes columnas:

#De esta manera agrupamos por year_of_award y hacemos un aggregate:
        #Para la columna age_at_award ponemos los dos puntos y luego la lista de operaciones que queremos (mean y count)
        #'year_of_birth':'min' --> para el año de nacimiento queremos el minimo.
        #'religion':['max','min'] --> y para religion el minimo y maximo.

#De esta manera con el group by aplicamos operaciones sobre 3 columnas distintas.   
grouped=oscars_short.groupby(['year_of_award']).agg({'age_at_award':['mean','count'],'year_of_birth':'min','religion':['max','min']}) 
grouped.head(10) #Hacemos un head(10 para mostrar solo los primeros 10 años... sino me mostraría toooodos)

#Vemos que tenemos un índice con 2 niveles:
    #Para age_at_Award: tenemos el mean y el count.
    #Para year_of_birth tenemos el min (enrealidad es la persona más joven que ganó premios ese año)
    #Y para religion tenemos el max y el min (para ver los valores de menor y mayor ocurrencia... si es el mismo seguramente es porque los valores son todos iguales para el mismo año)

#De esta manera tenemos estadísticas de 3 columnas distintas POR año de entrega del premio.
```

```python
#Como dijimos, cuando usamos groupby, inevitablemente crearemos un "multiindice" en las columnas o filas (o ambas).
#Y los DF con multiindices son dificultosos para navegar y podemos confundirnos en el nombre de las columnas.
#Ya que como vimos, obtenemos una estructura con 2 niveles de índices... que dificulta mucho el acceso a los datos.

#Por esto, vamos a querer "aplanar/simplificar" la estuctura... pero si quisieramos hacer esto tendríamos muchas columnas min y count... NO nos serviria esto.
#Entonces, lo que vamos a hacer es analizar el NIVEL de cada uno para poder unir los nombres:
```

```python
#En level 0 le asignamos el indice con el nivel 0:
level0=grouped.columns.get_level_values(0)

#Vemos en el gráfico de arriba que arriba de todo tenemos (Sería nuestro level 0, las columnas a las cuales realizamos las operaciones):
#	age_at_award	year_of_birth	religion
# y justo abajo (Sería nuestro level 1, serían las operaciones en si):
#	mean   count	 min	    max	 min
#Donde mean y count corresponden a age_at_award, min a year_of_birth, max y el 2do min a religion.

#De esta manera lo que llamamos nivel 0 es:
#age_at_award (por el mean), age_at_award (por el count), year_of_birth (por el min), religion (por el max) y religion (por el min)
```

```python
level0 #Imprmimimos para corroborar esto.
```

```python
#Y para level 1 tendremos como vimos anteriormente: mean, count, min, max y min nuevamente.
level1=grouped.columns.get_level_values(1)
```

```python
level1
```

```python
#Podemos ver que tenemos 2 columnas llamadas "min". Solo las podemos diferenciar usando ambos niveles de índices (0 y 1). 
#Entonces trataremos de reducir esta estructura concatenando ambos niveles en un solo nombre de columna:
grouped.columns = level0 + '_' + level1
#De esta manera tendremos nombre de columan (nivel 0) + _ + nombre operacion (nivel 1).
#Eso lo asignnamos para renombrar directamente los nombres de las columnas de grouped --> ".columns"
```

```python
#Haciendo esto, ahora tendremos 5 diferentes columna con UN SOLO NIVEL DE INDICES:
grouped

#Vemos que nos quedó muy bien:
#age_at_award_mean, age_at_award_count,year_of_birth_min, religion_max, religion_min.
#De esta manera es mucho más simple trabajar sobre las columnas... sino tendríamos que redireccionarnos con las tuplas. 
```

We can see that we still have 2 levels in the row index. We can reduce this using _reset_index()_

```python
#Pero todavia vemos que tenemos 2 niveles en el índice de FILA. Entonces usamos reset_index para 
#reiniciar el índice:
grouped.reset_index(inplace=True)
```

```python
#Ahora vamos a ver otra variante para aplicar transofmraciones de estadisticas sobre estructuras como ALTERNATIVA al group by: TRANSFORM.
oscars_short.head(3) #La estrcutura que teniamos era esta...
```

```python
#Supongamos que agrupamos por Award y luego sacamos el promedio de edades para CADA UNA de las distintas categorias:
grouped=oscars_short.groupby('award')
grouped['age_at_award'].mean()

#Entonces esto nos genera una estructura con 5 filas... para mejor actor el promedio es 45, para mejor actriz 36 y así (para las 5 categorías que aparecen).
```

```python
#Ahora si queremos comparar el promedio de la persona con el promedio de edad de su categoría...
#tenemos que hacer un JOIN de los datos anteriores con los datos de los premios para poder comparar.
#En vez de hacer eso hacemos un transform con el promedio:
grouped['age_at_award'].transform('mean').head(10)
#Este transform nos devuelve 1 elemento por cada una de las filas de nuestro DF...
#Vamos a tener tantos elementos como filas. Mostramos los primeros 20 pero tenemos los 400 y pico que teniamos.
```

```python
#Entonces a oscar_short lo agrupamos por premio --> .groupby('award') y le hacemos un transform del mean a la edad que tenía al recibir el premio --> ['age_at_award'].transform('mean')...
#Lo guardamos como una columna dentro de nuestro DF original --> oscars_short['avg_age_at_award'] =
#... Entonces para saber si la persona que gano el premio tenía más o menos edad que el promedio de su categoría lo que podemos hacer es comparar 2 valores dentro de nuestro mismo DF... que es más facil de hacer un MERGE que nos
#obligaba a hacer si hubiesemos usado groupby y apply que nos generaba una estructura de 5 datos... y con transform nos genera una columna entera con todos sus valores asociados (igual tamaño del registro del DF inicial)
oscars_short['avg_age_at_award'] = oscars_short.groupby('award')['age_at_award'].transform('mean')
#Para repetir... a la columna avg_age_at_award le asignamos (=) el resultado de agrupar por award y quedarnos con age_at_award y hacer un transform con el promedio (mean)... lo que nos generará una columna nueva donde para cada elemneto en función de su catogoría le asigna el promedio de edad...
#Así por ej. para los que tengan premio "Best Director" en avg_age_at_award le asignará 47.27907 a tooodos los que tengan ese premio. 

#Y por ultimo generamos una nueva columna is_above_avg_age que tendrá un booleano que dirá si su edad está sobre la edad promedio de su categoría (avg_age_at_award): 
oscars_short['is_above_avg_age'] = oscars_short['avg_age_at_award'] < oscars_short['age_at_award']
#Así, si el promedio es menor que la edad que tiene la persona esto va a ser T.
#Por ej. vemos la persona Frank Lloyd que tiene 48 años y el promedio para Best_Actor es 47... entonces en is_above_avg_agem es True!.

#De esta manera la salida del transform serán tantas filas como tenia nuestro DF originalmente. 
```
